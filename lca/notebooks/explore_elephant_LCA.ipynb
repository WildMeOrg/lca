```python
'''
Extract the log file lines.
'''
log_lines = []
fp = open(log_fn)
for line in fp:
    log_lines.append(line)
print(len(log_lines))
```

```python
'''
Get the initial edges that come from MIEWID
'''
log_str = 'ga_driver.py:278'     # Careful.  This may change if new code is inserted in ga_driver.py
init_edges = []
for line in log_lines:
    if log_str in line:
        i0 = line.find('(')
        i1 = line.find(')')
        wgt_substr = line[i0+1: i1]
        wgt_items = wgt_substr.split(',')
        a0 = int(wgt_items[0])
        a1 = int(wgt_items[1])
        w = int(wgt_items[2])
        init_edges.append((a0, a1, w))

print(len(init_edges))
print(init_edges[0])
print(init_edges[-1])
```

```python
#  √=INFO   2024-09-23 17:28:50,184 [            run.py:193] a0: 494, a1: 932, s:0.3885232210159302, w:-23
pr2sw = {}
log_str = 'run.py:203'
for line in log_lines:
    if log_str in line:
        i = line.find('a0:') + 4
        line = line[i:]
        i = line.find(',')
        a0 = int(line[:i])
        
        line = line[i+6:]
        i = line.find(',')
        a1 = int(line[:i])
        
        line = line[i+4:]
        i = line.find(',')
        s = float(line[:i])

        w = int(line[i+4:])

        pr2sw[(a0, a1)] = (s, w)

for a0, a1, w in init_edges:
    assert(pr2sw[(a0, a1)][1] == w)

print(len(pr2sw))
```

```python
pr2wgt = {pr:pr2sw[pr][1] for pr in pr2sw.keys()}
```

```python
'''
Form a weight histogram for the initial edges passed to LCA to get it started.
'''
bins = [0]*20
for _, _, w in init_edges:
    wi = (w + 100) // 10
    wi = min(wi, 19)
    bins[wi] += 1

print('Histogram of weights')
for i in range(20):
    wmin = i*10 - 100
    wmax = wmin+10
    print(f'{wmin}, {wmax}: {bins[i]}')

print(f'Num negative weights {sum(bins[0:10])}')
print(f'Num positive weights {sum(bins[10:])}')
```

```python
'''
Get the mapping from annotation to cluster
'''
a2c = dict()
for cid, annots in gt_clustering.items():
    for a in annots:
        a2c[a] = cid
# print(a2c)
print(len(a2c))
print(a2c[1120])
```

```python
'''
Form a dictionary mapping annotations to their list of edge neighbors
'''
from collections import defaultdict
a2neighbors = defaultdict(list)
for a0, a1 in pr2wgt:
    a2neighbors[a0].append(a1)
    a2neighbors[a1].append(a0)
print(len(a2neighbors))
```

```python
'''
For each ground truth cluster, find:
1. The fraction of correct edges that are there
2. The number of incorrect edges
3. The min, max, avg and std deviations of the scores for internal and external edges
# 4. The same for weights

Use NetworkX to determine the number of components of each cluster


'''
import networkx as nx
import sympy

num_components = 0
num_connected = 0
num_edges_within = 0
num_edges_between = 0

for cid, annots in gt_clustering.items():
    total_possible_correct = sympy.binomial(len(annots), 2)

    internal_s = []
    internal_w = []
    external_s = []
    external_w = []

    G = nx.Graph()
    G.add_nodes_from(annots)
    
    for a in annots:
        for a1 in a2neighbors[a]:
            if a1 not in annots:
                a0 = min(a, a1)
                a1 = max(a, a1)
                s, w = pr2sw[(a0, a1)]
                external_s.append(s)
                external_w.append(w)
            elif a < a1:
                s, w = pr2sw[(a, a1)]
                internal_s.append(s)
                internal_w.append(w)
                G.add_edge(a, a1)

    num_edges_within += len(internal_s)
    num_edges_between += len(external_s) // 2  # avoid double counting
    
    print(f'----------\nCluster {cid}, length {len(annots)}')
    print(f'Total possible internal edges {total_possible_correct}')
    print(f'Actual number of internal edges {len(internal_s)}')
    if len(external_s) > 0:
        print(f'External scores:  min {min(external_s)}, max {max(external_s)}, avg {sum(external_s)/len(external_s)}')
        print(f'External weights:  min {min(external_w)}, max {max(external_w)}, avg {sum(external_w)/len(external_w)}')
    else:
        print(f'External scores: None')

    if len(internal_s) > 0:
        print(f'Internal scores:  min {min(internal_s)}, max {max(internal_s)}, avg {sum(internal_s)/len(internal_s)}')
        print(f'Internal weights:  min {min(internal_w)}, max {max(internal_w)}, avg {sum(internal_w)/len(internal_w)}')
    else:
        print(f'Internal scores: None')

    cc = list(nx.connected_components(G))
    num_components += len(cc)
    if len(cc) <= 1:
        print('Internally connected')
        num_connected += 1
    else:
        print(f'Internally disconnected into {len(cc)} components: {cc}')
```

```python
print('===========')
print(f'Number of ground truth clusters {len(gt_clustering)}')
print(f'Number of internally connected GT clusters {num_connected}. This fraction is {num_connected / len(gt_clustering)}')
print(f'Number of clusters within the GT clustering {num_components}')
print(f'Number of edges within GT clusters {num_edges_within}')
print(f'Number of edges between GT clusters {num_edges_between}')
```
